```python
from mpl_toolkits import mplot3d
x = np.linspace(10,70,20)
y=np.linspace(10,70,20)
y
```

```python
z = x+y
def sin_fun(x,y):
    return np.sin(np.sqrt(x**2+y**2))
z = sin_fun(x,y)
ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z)
```

```python
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel("z")
```

```python
ax = plt.axes(projection = '3d')
ax.plot3D(df['petal_width'],df['sepal_width'])
ax.set_ylabel('petal_length')
ax.set_zlabel("sepal_width")
```
